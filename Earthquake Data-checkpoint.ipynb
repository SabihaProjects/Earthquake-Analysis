 Earthquake Analysis

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = pd.read_csv("Earthquake Data.csv")
a
```

```python
# Q1 How to plot the top 10 places with most earthquakes?
top_places = a['place'].value_counts().head(10)
top_places.plot(kind='bar', color='orange')
plt.title("Top 10 Earthquake Places")
plt.xlabel("Place")
plt.ylabel("Number of Earthquakes")
plt.show()

```

```python
#Q2 How to visualize the distribution of earthquake magnitudes?
a['mag'].plot(kind='hist', bins=20, color='skyblue')
plt.title("Distribution of Earthquake Magnitudes")
plt.xlabel("Magnitude")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q3 How to analyze the depth of earthquakes?
a['depth'].plot(kind='hist', bins=20, color='green')
plt.title("Distribution of Earthquake Depths")
plt.xlabel("Depth (km)")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q4 How to compare magnitude vs depth of earthquakes?
a.plot(kind='scatter', x='depth', y='mag', color='red')
plt.title("Magnitude vs Depth")
plt.xlabel("Depth (km)")
plt.ylabel("Magnitude")
plt.show()
```

```python
#Q5 How to see the monthly trend of earthquakes?
a['Month'] = pd.to_datetime(a['time']).dt.month
monthly_counts = a['Month'].value_counts().sort_index()
monthly_counts.plot(kind='line', marker='o', color='purple')
plt.title("Earthquakes per Month")
plt.xlabel("Month")
plt.ylabel("Number of Earthquakes")
plt.show()
```

```python
#Q6 How to categorize earthquakes by magnitude?
bins = [0, 3, 5, 7, 10]
labels = ['Small', 'Moderate', 'Strong', 'Major']
categories = pd.cut(a['mag'], bins=bins, labels=labels)
categories.value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['yellow','orange','red','darkred'])
plt.title("Earthquakes by Magnitude Category")
plt.ylabel("")
plt.show()
```

```python
#Q7 How to visualize earthquake locations on a map?
plt.scatter(a['longitude'], a['latitude'], c=a['mag'], cmap='viridis', alpha=0.6)
plt.colorbar(label='Magnitude')
plt.title("Earthquake Locations by Magnitude")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
#Q8 How to check relationship between RMS and magnitude?
a.plot(kind='scatter', x='rms', y='mag', color='brown')
plt.title("RMS vs Magnitude")
plt.xlabel("RMS")
plt.ylabel("Magnitude")
plt.show()
```

```python
#Q9 How to check the distribution of gap values?
a['gap'].plot(kind='hist', bins=20, color='pink')
plt.title("Distribution of Gap Values")
plt.xlabel("Gap")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q10 How to check number of earthquakes by type?
a['type'].value_counts().plot(kind='bar', color='cyan')
plt.title("Number of Earthquakes by Type")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```
